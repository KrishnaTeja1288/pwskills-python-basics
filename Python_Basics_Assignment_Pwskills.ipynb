```python
# 3. Compare and contrast mutable and immutable objects in Python with examples


'''

In Python, mutable and immutable objects refer to whether the state (i.e., value) of an object can be changed after it is created. Here's a comparison of mutable and immutable objects, along with examples to illustrate their differences.

 Mutable Objects
- Definition: Mutable objects are those whose values or state can be modified after creation.
- Examples: Lists, dictionaries, sets, and some user-defined classes.

# Characteristics:
1. Changeable: You can alter the content of mutable objects (add, remove, or change elements) without creating a new object.
2. Memory Efficiency: Since changes can happen in-place, modifying mutable objects doesn't necessarily create new objects, saving memory in some cases.
3. Methods: Mutable objects provide methods that allow modifications (e.g., append(), remove() for lists, update() for dictionaries).

# Example (List - a mutable object):

my_list = [1, 2, 3]
print(f"Original list: {my_list}")

my_list[1] = 20  # Modify the list
print(f"Modified list: {my_list}")

Output:

Original list: [1, 2, 3]
Modified list: [1, 20, 3]

Here, the value at index 1 was changed from 2 to 20 without creating a new list.

# Another Example (Dictionary - a mutable object):

my_dict = {'name': 'Alice', 'age': 25}
print(f"Original dictionary: {my_dict}")

my_dict['age'] = 26  # Modify the dictionary
print(f"Modified dictionary: {my_dict}")

Output:

Original dictionary: {'name': 'Alice', 'age': 25}
Modified dictionary: {'name': 'Alice', 'age': 26}

Here, the value associated with the key 'age' was changed from 25 to 26.

 Immutable Objects
- Definition: Immutable objects are those whose values cannot be changed after they are created. Any attempt to modify an immutable object will result in the creation of a new object.
- Examples: Integers, floats, strings, tuples, and frozensets.

# Characteristics:
1. Unchangeable: Once created, you cannot change the value or state of immutable objects. Any change will result in a new object being created.
2. Memory Usage: Every modification creates a new object, which may increase memory usage if not handled properly.
3. Hashable: Immutable objects can be used as keys in dictionaries or elements of sets, whereas mutable objects cannot (since their value could change, leading to inconsistencies).

# Example (String - an immutable object):

my_string = "hello"
print(f"Original string: {my_string}")

my_string = my_string + " world"  # Concatenation creates a new string
print(f"Modified string: {my_string}")

Output:

Original string: hello
Modified string: hello world

Here, my_string was modified, but instead of changing the original string, a new string "hello world" was created.

# Another Example (Tuple - an immutable object):

my_tuple = (1, 2, 3)
print(f"Original tuple: {my_tuple}")

# Attempting to modify the tuple will raise an error
# my_tuple[0] = 10  # Uncommenting this will cause an error

Output:

Original tuple: (1, 2, 3)

Tuples cannot be modified, and attempting to do so will result in an error.

 Key Differences Between Mutable and Immutable Objects

| Aspect                 | Mutable Objects                     | Immutable Objects                   |
||--|-|
| Modifiability       | Can be modified after creation       | Cannot be modified after creation   |
| Examples            | List, dictionary, set               | String, tuple, integer, float       |
| Memory Behavior     | Changes happen in place, saving memory | New objects are created for changes |
| Hashable            | Not hashable (can’t be used as dictionary keys) | Hashable (can be used as keys) |
| Use Cases           | Useful when frequent modifications are needed | Useful when the object should remain constant |

 Example of Mutable vs. Immutable Behavior in Functions:

# Mutable (List):

def modify_list(my_list):
    my_list.append(4)

lst = [1, 2, 3]
modify_list(lst)
print(f"List after modification: {lst}")

Output:

List after modification: [1, 2, 3, 4]

The list is modified inside the function, and the change is reflected outside because lists are mutable.

# Immutable (Integer):

def modify_number(num):
    num = num + 1

number = 5
modify_number(number)
print(f"Number after modification: {number}")

Output:

Number after modification: 5

The integer remains unchanged because integers are immutable, and modifying it inside the function results in a new object.

 Conclusion:
- Mutable objects allow changes after creation and are useful when you need to update the object frequently.
- Immutable objects cannot be changed and are often used when consistency, reliability, or memory efficiency is required.

Understanding the difference between mutable and immutable objects is important for writing efficient and bug-free Python code.
'''
```

```python
#  4. Discuss the different types of operators in Python and provide examples of how they are used

'''
In Python, operators are special symbols or keywords that perform operations on variables and values. Python has several types of operators, each used for different purposes. Here's a discussion of the main types of operators along with examples of how they are used.

 1. Arithmetic Operators
Arithmetic operators are used to perform mathematical operations like addition, subtraction, multiplication, division, etc.

# Example:

a = 10
b = 3

print(a + b)    # Addition -> 13
print(a - b)    # Subtraction -> 7
print(a * b)    # Multiplication -> 30
print(a / b)    # Division -> 3.333...
print(a % b)    # Modulus -> 1 (remainder of 10 / 3)
print(a  b)   # Exponentiation -> 1000 (10 raised to the power of 3)
print(a // b)   # Floor Division -> 3 (quotient of 10 / 3 without remainder)


 2. Comparison (Relational) Operators
Comparison operators are used to compare two values and return a Boolean (True or False) depending on the result.


# Example:

a = 10
b = 5

print(a == b)   # False (10 is not equal to 5)
print(a != b)   # True (10 is not equal to 5)
print(a > b)    # True (10 is greater than 5)
print(a < b)    # False (10 is not less than 5)
print(a >= b)   # True (10 is greater than or equal to 5)
print(a <= b)   # False (10 is not less than or equal to 5)


 3. Logical Operators
Logical operators are used to combine conditional statements and return True or False.

# Example:

x = True
y = False

print(x and y)  # False (since one of them is False)
print(x or y)   # True (since one of them is True)
print(not x)    # False (negates True to False)


 4. Assignment Operators
Assignment operators are used to assign values to variables and perform operations like addition, subtraction, etc., while assigning the result to the variable.

# Example:

a = 10
a += 5  # Equivalent to a = a + 5
print(a)  # 15

a *= 2  # Equivalent to a = a * 2
print(a)  # 30


 5. Bitwise Operators
Bitwise operators work on bits (binary representations of numbers) and perform operations like AND, OR, XOR, etc.

# Example:

a = 60  # Binary: 0011 1100
b = 13  # Binary: 0000 1101

print(a & b)   # Bitwise AND -> 12 (Binary: 0000 1100)
print(a | b)   # Bitwise OR -> 61 (Binary: 0011 1101)
print(a ^ b)   # Bitwise XOR -> 49 (Binary: 0011 0001)
print(~a)      # Bitwise NOT -> -61 (inverts bits)
print(a << 2)  # Left shift -> 240 (Binary: 1111 0000)
print(a >> 2)  # Right shift -> 15 (Binary: 0000 1111)


 6. Membership Operators
Membership operators are used to test whether a value is a member of a sequence like a list, tuple, or string.

# Example:

fruits = ["apple", "banana", "cherry"]

print("banana" in fruits)    # True (banana is in the list)
print("grape" not in fruits) # True (grape is not in the list)


 7. Identity Operators
Identity operators compare the memory locations of two objects, determining if they are the same object or not.

# Example:

a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)    # True (a and b refer to the same object)
print(a is c)    # False (a and c refer to different objects)
print(a == c)    # True (a and c have the same values)


 8. Special Operators

# Ternary Operator (Conditional Expression):
It’s a one-liner decision-making statement.


a = 10
b = 20
min_value = a if a < b else b
print(min_value)  # 10 (because a is smaller than b)


 Summary of Python Operators:
- Arithmetic operators perform basic mathematical operations.
- Comparison operators compare two values.
- Logical operators combine multiple conditions.
- Assignment operators assign and modify values in place.
- Bitwise operators perform operations on binary representations of numbers.
- Membership operators check for membership in a sequence.
- Identity operators compare objects' memory locations.

'''
```

```python
# 5.  Explain the concept of type casting in Python with examples


'''
Ans)
Type casting in Python refers to the process of converting a variable from one data type to another. Python supports both implicit and explicit type casting. Let’s explore both concepts with examples.

1. Implicit Type Casting
In implicit type casting, Python automatically converts one data type to another without user intervention. This usually occurs when combining compatible types (like integers and floats) to avoid data loss.

 Example:
x = 5     # Integer
y = 2.5   # Float

result = x + y  # x is implicitly cast to float
print(result)   # Output: 7.5
print(type(result))  # Output: <class 'float'>

Here, Python converts x (an integer) to a float to match the data type of y during the addition.



 2. Explicit Type Casting
Explicit type casting (or type conversion) is when the programmer manually converts one data type into another using built-in functions like int(), float(), str(), and more.

 Example 1: Converting Float to Integer
python
num = 3.7
int_num = int(num)  # Explicit type casting
print(int_num)  # Output: 3
print(type(int_num))  # Output: <class 'int'>

Here, int() discards the decimal part and converts the float to an integer.

 Example 2: Converting String to Integer
python
num_str = "100"
num = int(num_str)  # Explicitly cast string to integer
print(num)  # Output: 100
print(type(num))  # Output: <class 'int'>

If the string is not a valid number (e.g., "abc"), this will raise a ValueError.

 Example 3: Converting Integer to String
python
age = 25
age_str = str(age)  # Convert integer to string
print(age_str)  # Output: '25'
print(type(age_str))  # Output: <class 'str'>




 Summary of Common Type Casting Functions:
| Function   | Description                            |
||-|
| int()    | Converts to an integer                |
| float()  | Converts to a float                   |
| str()    | Converts to a string                  |
| list()   | Converts to a list                    |
| tuple()  | Converts to a tuple                   |
| set()    | Converts to a set                     |
| bool()   | Converts to a boolean (True/False)  |



By using implicit and explicit type casting effectively, you can ensure your program behaves as expected when working with multiple data types.
'''
```

```python
# 7.  Describe the different types of loops in Python and their use cases with examples.?
'''
Ans)

In Python, loops allow us to execute a block of code repeatedly. This helps automate repetitive tasks efficiently. Python has two main types of loops:

1. for loop
2. while loop

Additionally, Python supports nested loops and provides control flow statements like break, continue, and pass to manage the behavior of loops.



 1. for Loop

A for loop is used to iterate over a sequence (like a list, tuple, string, or range). It repeats the block of code for each element in the sequence.

 Syntax:
python
for variable in sequence:
    # Code to execute for each element


 Example 1: Looping over a List
python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

Output:

apple
banana
cherry


 Example 2: Using range() in a for Loop
The range() function generates a sequence of numbers. It's often used to loop over a specific number of times.

python
for i in range(5):
    print(i)

Output:

0
1
2
3
4

Here, the loop runs 5 times, from 0 to 4.

 Use Case:
for loops are useful when you know in advance how many times you want to iterate (e.g., iterating over a list or running a loop a fixed number of times).



 2. while Loop

A while loop repeats a block of code as long as a condition is True. If the condition becomes False, the loop stops.

 Syntax:
python
while condition:
    # Code to execute while condition is True


 Example: Printing Numbers Until a Condition is Met
python
i = 1
while i <= 5:
    print(i)
    i += 1

Output:

1
2
3
4
5


 Use Case:
while loops are useful when you don't know the exact number of iterations in advance (e.g., waiting for a certain condition to become True).



3. Nested Loops

A nested loop is a loop inside another loop. It allows you to iterate over multiple dimensions (like matrices).

 Example: Printing a Pattern
python
for i in range(3):
    for j in range(3):
        print(f"({i}, {j})", end=" ")
    print()

Output:

(0, 0) (0, 1) (0, 2)
(1, 0) (1, 1) (1, 2)
(2, 0) (2, 1) (2, 2)




 4. Loop Control Statements

 1. break:
The break statement stops the loop immediately when a certain condition is met.

python
for i in range(10):
    if i == 5:
        break
    print(i)

Output:

0
1
2
3
4

Here, the loop stops when i becomes 5.

 2. continue:
The continue statement skips the current iteration and moves to the next one.

python
for i in range(5):
    if i == 2:
        continue
    print(i)

Output:

0
1
3
4


 3. pass:
The pass statement does nothing. It is used as a placeholder for code that you plan to write later.

python
for i in range(3):
    pass  # Placeholder for future code




 Use Cases for Different Loops

- for loop:
  - Iterating over elements of a list, tuple, or dictionary.
  - Running a loop a fixed number of times using range().

- while loop:
  - Repeating actions until a condition becomes False (e.g., waiting for user input to be valid).

- Nested loops:
  - Handling multi-dimensional data structures (like matrices or grids).

- Loop control statements:
  - break: Stopping the loop based on a condition.
  - continue: Skipping unwanted iterations.
  - pass: Acting as a placeholder for future code.

'''
```
